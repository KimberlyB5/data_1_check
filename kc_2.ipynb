```python
#To aquire the columns I want to use
df = pd.read_csv("C:\\Users\\kam52\\OneDrive\\Desktop\\DA1\\data_1_check\\Monkeypox.csv", usecols = ['Country','Status'])
print(df)
```

```python
#Renaming column Status to Cases for the count()
new_name = {'Country' : 'Country', 'Status' : 'Cases'}

dx = df.rename(columns = new_name)
print(dx)
```

```python
# To find how many Countries are listed
Origin = pd.unique(dx['Country'])
print(Origin)
```

```python
#To count the number of cases per country
country_group= dx.groupby('Country').count()
print(country_group)
country_group.dtypes
```

```python
#plot the graph

country_group.plot (kind ='bar', figsize=(12,6))
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.show()
```

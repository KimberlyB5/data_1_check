```python
#To count how many sightings per state
def state():
    ds = pd.read_csv(os.path.join("https://query.data.world/s/ti4reepebaxvozsotydx6vm3nyqz3u"),
        usecols = ['city', 'state'])
    state_city= ds.groupby('state').count()
    return state_city
  
state()
```

```python
#To plot data from state() on graph

def state_graph():
    ds = pd.read_csv(os.path.join("https://query.data.world/s/ti4reepebaxvozsotydx6vm3nyqz3u"),
        usecols = ['city', 'state'])
    state_city= ds.groupby('state').count()
    state_city.plot (kind ='bar', figsize=(18,9))
    plt.xlabel('State or Country')
    plt.ylabel('Sightings Reported')
    
state_graph()
```
